```python
import os
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, GlobalAveragePooling2D, Dense, Dropout, BatchNormalization, Flatten
from tensorflow.keras.regularizers import l2
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import LearningRateScheduler
from tensorflow.keras.preprocessing.image import ImageDataGenerator 
```

```python
import os
from PIL import Image
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Paths for train, validation, and test data directories
train_data_dir = 'data/to/train/folder'
validation_data_dir = 'data/to/validation/folder'
test_data_dir = 'data/to/test/folder'

# Function to validate images and remove corrupted ones
def validate_images(directory):
    for root, _, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)
            try:
                img = Image.open(file_path)
                img.verify()  
            except (IOError, SyntaxError) as e:
                print(f"Corrupted file: {file_path}")
                os.remove(file_path) 

# Validate all datasets
validate_images(train_data_dir)
validate_images(validation_data_dir)
validate_images(test_data_dir)

# Image data generators for training, validation, and test datasets
train_datagen = ImageDataGenerator(
    rescale=1.0/255.0,  # Rescale images
    rotation_range=30,
    width_shift_range=0.2, 
    height_shift_range=0.2,
    shear_range=0.2, 
    zoom_range=0.2, 
    horizontal_flip=True, 
    fill_mode='nearest'
)

validation_datagen = ImageDataGenerator(rescale=1.0/255.0)  # Only rescaling for validation

test_datagen = ImageDataGenerator(rescale=1.0/255.0)  # Only rescaling for test

# Training data generator
train_generator = train_datagen.flow_from_directory(
    train_data_dir,
    target_size=(128, 128),  # Resizing images to 128x128
    batch_size=32,
    class_mode='sparse'  # Assuming labels are sparse integers
)

# Validation data generator
validation_generator = validation_datagen.flow_from_directory(
    validation_data_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode='sparse'  # Same assumption for validation
)

# Test data generator
test_generator = test_datagen.flow_from_directory(
    test_data_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode='sparse'  # Same assumption for test
)
```

```python
for batch_data, batch_labels in train_generator:
    print(f"Data shape: {batch_data.shape}, Labels shape: {batch_labels.shape}")
    break
```

```python
# 1 St  architecture for model in 4 blocks

model = Sequential()

model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)))
model.add(MaxPooling2D((2, 2)))
# Second Convolutional Layer
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
# Flatten Layer
model.add(Flatten())
# Fully Connected Layer
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))  # Dropout to reduce overfitting
# Output Layer
model.add(Dense(6, activation='softmax'))


# 2nd architecture for model in 4 blocks

# model.add(Conv2D(32, (3, 3), activation='relu', kernel_regularizer=l2(0.01), input_shape=(128, 128, 3)))
# model.add(BatchNormalization())
# model.add(MaxPooling2D((2, 2)))
# model.add(Dropout(0.3))

# model.add(Conv2D(64, (3, 3), activation='relu', kernel_regularizer=l2(0.01)))
# model.add(BatchNormalization())
# model.add(MaxPooling2D((2, 2)))
# model.add(Dropout(0.3))

# model.add(Conv2D(128, (3, 3), activation='relu', kernel_regularizer=l2(0.01)))
# model.add(BatchNormalization())
# model.add(MaxPooling2D((2, 2)))
# model.add(Dropout(0.4))

# model.add(Conv2D(256, (3, 3), activation='relu', kernel_regularizer=l2(0.01)))
# model.add(BatchNormalization())
# model.add(MaxPooling2D((2, 2)))
# model.add(Dropout(0.4))

# model.add(GlobalAveragePooling2D())

# model.add(Dense(256, activation='relu', kernel_regularizer=l2(0.01)))
# model.add(BatchNormalization())
# model.add(Dropout(0.5))
# model.add(Dense(128, activation='relu', kernel_regularizer=l2(0.01)))
# model.add(BatchNormalization())
# model.add(Dropout(0.5))

# num_classes = len(train_generator.class_indices)
# model.add(Dense(num_classes, activation='softmax'))

# 3rd architecture for model in 4 blocks

# model.add(Conv2D(32, (3, 3), activation='relu', kernel_regularizer=l2(0.01), input_shape=(128, 128, 3)))
# model.add(BatchNormalization())
# model.add(MaxPooling2D((2, 2)))

# model.add(Conv2D(64, (3, 3), activation='relu', kernel_regularizer=l2(0.01)))
# model.add(BatchNormalization())
# model.add(MaxPooling2D((2, 2)))

# model.add(Conv2D(128, (3, 3), activation='relu', kernel_regularizer=l2(0.01)))
# model.add(BatchNormalization())
# model.add(MaxPooling2D((2, 2)))

# model.add(GlobalAveragePooling2D())

# model.add(Dense(128, activation='relu', kernel_regularizer=l2(0.01)))
# model.add(BatchNormalization())
# model.add(Dropout(0.5))

# num_classes = len(train_generator.class_indices)  # Number of output classes
# model.add(Dense(num_classes, activation='softmax'))

# 4rd architecture for model in 4 blocks

# model.add(Conv2D(32, (3, 3), input_shape=(128, 128, 3)))
# model.add(BatchNormalization())
# model.add(MaxPooling2D((2, 2)))
# model.add(Dropout(0.3))  # Dropout in Conv Block

# model.add(Conv2D(64, (3, 3)))
# model.add(BatchNormalization())
# model.add(MaxPooling2D((2, 2)))
# model.add(Dropout(0.3))  # Dropout in Conv Block

#  Global Pooling (Replaces Flatten)
# model.add(GlobalAveragePooling2D())

# model.add(Dense(128, activation='relu'))
# model.add(Dropout(0.5))  # Dropout for dense layer

# num_classes = len(train_generator.class_indices)  # Replace with your number of classes
# model.add(Dense(6, activation='softmax'))



def lr_scheduler(epoch, lr):
    if epoch % 10 == 0 and epoch > 0:
        return lr * 0.5
    return lr

model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()


callbacks = [
    LearningRateScheduler(lr_scheduler)
]


history = model.fit(
    train_generator,
    steps_per_epoch=(train_generator.samples + train_generator.batch_size - 1) // train_generator.batch_size,
    epochs=30,
    validation_data=validation_generator,
    validation_steps=(validation_generator.samples + validation_generator.batch_size - 1) // validation_generator.batch_size,
    callbacks=callbacks
)
```
